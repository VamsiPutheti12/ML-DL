```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Initialize parameters
w = 0.0
b = 0.0
lr = 0.01
epochs = 20

# Track metrics
losses = []
w_vals = []
b_vals = []

# SGD Training Loop
for epoch in range(epochs):
    # Shuffle data
    indices = np.arange(len(X))
    np.random.shuffle(indices)

    # Loop through each shuffled sample
    for i in indices:
        xi = X[i]
        yi = y[i]

        # Prediction
        y_pred = w * xi + b
        error = y_pred - yi

        # Gradients
        dw = 2 * xi * error
        db = 2 * error

        # Parameter update
        w -= lr * dw
        b -= lr * db

    # Calculate and store full-batch loss after this epoch
    y_preds = w * X + b
    loss = np.mean((y_preds - y) ** 2)
    losses.append(loss)
    w_vals.append(w)
    b_vals.append(b)

# Final model
print(f"\nFinal model: y = {w:.4f}x + {b:.4f}")

#  Plotting
plt.figure(figsize=(12, 4))

# Loss Plot
plt.subplot(1, 3, 1)
plt.plot(range(1, epochs + 1), losses, marker='o')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True)

# Weight Plot
plt.subplot(1, 3, 2)
plt.plot(range(1, epochs + 1), w_vals, marker='x', color='green')
plt.title('Weight (w) Convergence')
plt.xlabel('Epoch')
plt.ylabel('w')
plt.grid(True)

# Bias Plot
plt.subplot(1, 3, 3)
plt.plot(range(1, epochs + 1), b_vals, marker='s', color='orange')
plt.title('Bias (b) Convergence')
plt.xlabel('Epoch')
plt.ylabel('b')
plt.grid(True)

plt.tight_layout()
plt.show()
```
